# Overall overview of the project


- The project aims to show the capability of using Named Entity Recognition Models to find specific structured data contained in  semi-structured data. 
- The NER model in this project is trained to find specific entities such as ingredients, cookware, measurements, temperature, etc. in the recipe text.
- The overall goal of this project is to make searching for recipes easier, especially when consumer's ability to cook are very limited by the ingredients, cookware, and kitchen specs that are available to them.

# Challenges & Improvements

## DataParser

Challenges & Solutions:
- Did not know how to make scrapy to follow links within pages
    - Decided to use two different spiders to solve the issue
- Kept on forgetting the different scrapy cli
    - Decided to create file named scrapy.md that contains a mini-documentation for the framework to help make the project run smoother
- The extracted links and recipes did not equal to each other
    - Created Validation.py to find out why they were not equal

Recommendation/s:

- Combining the spiders link and recipes spiders to make it more organized. This can be achieved by applying scrapy CrawlSpider.
- Extracting data from more than one website, since the model is having a hard time to predict temperature and measurement values that uses the metric system.
- Creating a more specific preprocessing algorithm, since the preprocessing algorithm used in DataParser was too in-general when removing punctuations from the paragraph.
- Optimizing the data structure of the output data primarily removing unnecessary list and change the method from .getall() to .get().
- Learn more about scrapy overall such as what the none spider and setting python files do.
